### Part 2: Delving into Product Pricing

**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

3. **Visualizations**:
    - Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.

```python
# **Measures of Centrality**:
# Calculate the mean, median, and mode for the `price` of products.
# What's the average price point of products listed? How does this compare with the most common price point (mode)?
```

```python
# Calculate the mean, median, and mode for the `price` of products.
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()
```

```python
# What's the average price point of products listed? How does this compare with the most common price point (mode)?
mean_price, median_price, mode_price
```

```python
# **Measures of Dispersion**:
# Determine the variance, standard deviation, range, and interquartile range for product `price`.
# How varied are the product prices? Are there any indicators of a significant spread in prices?
```

```python
# Determine the variance, standard deviation, range, and interquartile range for product `price`.
price_variance = df['price'].var()
price_std = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)
```

```python
# How varied are the product prices? Are there any indicators of a significant spread in prices?
price_variance, price_std, price_range, price_iqr
```

```python
# 3. **Visualizations**:
# Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
# Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.
```

```python
# Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices.
plt.figure(figsize=(10, 20))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
# *If its hard to read these diagrams, think why this is, and explain how it could be solved.*
# The histogram is hard to read because of the presence of outliers. We can solve this by removing the outliers and plotting the histogram again.
# remove the outlier and again plot
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.xlim(0, 500)
plt.show()
```

```python
# Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='price')
plt.title('Spread of Product Prices')
plt.ylabel('Price')
plt.show()
```

### Part 3: Unpacking Product Ratings

**Business Question**: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `rating` of products.
    - How do customers generally rate products? Is there a common trend?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, and interquartile range for product `rating`.
    - Are the ratings consistent, or is there a wide variation in customer feedback?

3. **Shape of the Distribution**:
    - Calculate the skewness and kurtosis for the `rating` column. 
    - Are the ratings normally distributed, or do they lean towards higher or lower values?

4. **Visualizations**:
    - Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
# **Measures of Centrality**:
# Calculate the mean, median, and mode for the `rating` of products.
# How do customers generally rate products? Is there a common trend?
```

```python
# Calculate the mean, median, and mode for the `rating` of products.
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()
```

```python
# How do customers generally rate products? Is there a common trend?
mean_rating, median_rating, mode_rating
```

```python
# **Measures of Dispersion**:
# Determine the variance, standard deviation, and interquartile range for product `rating`.
# Are the ratings consistent, or is there a wide variation in customer feedback?
```

```python
# Measures of Dispersion**:
# Determine the variance, standard deviation, and interquartile range for product `rating`.
rating_variance = df['stars'].var()
rating_std = df['stars'].std()
rating_iqr = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

```

```python
# Are the ratings consistent, or is there a wide variation in customer feedback?
rating_variance, rating_std, rating_iqr
```

```python
# Shape of the Distribution**:
# Calculate the skewness and kurtosis for the `rating` column. 
# Are the ratings normally distributed, or do they lean towards higher or lower values?
```

```python
# Calculate the skewness and kurtosis for the `rating` column.
rating_skew = df['stars'].skew()
rating_kurt = df['stars'].kurt()
```

```python
# Are the ratings normally distributed, or do they lean towards higher or lower values?
rating_skew, rating_kurt
# what is the analysis of the above values
# The skewness of the ratings is -1.27, which indicates that the ratings are negatively skewed, i.e., they lean towards higher values.
# The kurtosis of the ratings is 2.03, which indicates that the ratings are platykurtic, i.e., they have fewer extreme values than a normal distribution.

```

```python
#  **Visualizations**:
# Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?
# Use a box plot to visualize the spread of product ratings. Are there any outliers in the data?
```

```python
# Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], bins=9, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```

```python
# Use a box plot to visualize the spread of product ratings. Are there any outliers in the data?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='stars')
plt.title('Spread of Product Ratings')
plt.ylabel('Rating')
plt.show()
```
